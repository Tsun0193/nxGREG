```python
import glob
import os
import shutil

root = "ctc-data-translated"
pattern = "*-en.md"
queries_pattern = "*sql-queries.md"
statements_pattern = "*sql-statements.md"
patterns = [pattern, queries_pattern, statements_pattern]

# print(files)
for pattern in patterns:
    files = glob.glob(os.path.join(root, "**", pattern), recursive=True)
    for f in files:
        dst_root = "ctc-data-en"
        new_path = f.replace(root, dst_root)
        print(os.path.dirname(new_path))
        # print(new_path)
        
        os.makedirs(os.path.dirname(new_path), exist_ok=True)
        shutil.copy(f, new_path)


```
