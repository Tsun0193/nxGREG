```python
from neo4j import GraphDatabase
from dotenv import load_dotenv
import os
load_dotenv()


uri = os.getenv("NEO4J_URL")
print(uri)
driver = GraphDatabase.driver(uri, auth=(os.getenv("NEO4J_USERNAME"),
                                         os.getenv("NEO4J_PASSWORD")))

def test_connection(tx):
    result = tx.run("RETURN 'Connected to Neo4j' AS msg")
    return result.single()["msg"]

with driver.session() as session:
    print(session.execute_write(test_connection))
    
    
```

```python
import glob
import os
import shutil

root = "ctc-data-translated"
pattern = "*-en.md"
files = glob.glob(os.path.join(root, "**", pattern), recursive=True)

# print(files)
for f in files:
    dst_root = "ctc-data-en"
    new_path = f.replace(root, dst_root)
    print(os.path.dirname(new_path))
    # print(new_path)
    
    os.makedirs(os.path.dirname(new_path), exist_ok=True)
    shutil.copy(f, new_path)

```
